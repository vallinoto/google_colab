```
import pandas as pd

pd.set_option("display.max_columns", 200)
```

```
TRAIN_FILE = "https://raw.githubusercontent.com/vallinoto/challenges/master/codenation/data/train.csv"
TEST_FILE  = "https://raw.githubusercontent.com/vallinoto/challenges/master/codenation/data/test.csv"
```

```
enem_train = pd.read_csv(TRAIN_FILE, sep=',', encoding='utf-8')
enem_test = pd.read_csv(TEST_FILE, sep=',', encoding='utf-8')
```

```
from pandas_profiling import ProfileReport

 
profile = ProfileReport(enem_train)
```

```
profile.to_file(output_file='output.html')
```

```
correlacao_features = enem_train.corr()
```

```
correlacao_features
```

```
features_selecionadas = correlacao_features[(correlacao_features['NU_NOTA_MT'] <= -0.28) | 
                       (correlacao_features['NU_NOTA_MT'] >= 0.28) & 
                       (correlacao_features['NU_NOTA_MT'] < 1.0)]['NU_NOTA_MT']
features_selecionadas = features_selecionadas.index.to_list()
features_selecionadas.append('NU_NOTA_MT')
```

```
enem_train[features_selecionadas]
```

```
from pycaret.regression import *


regressor = setup(
    data = enem_train[features_selecionadas], 
    target = 'NU_NOTA_MT', 
    feature_selection=True,
    feature_selection_threshold= 0.7,
    ignore_low_variance=True,
    silent=True

)
```

```
best = compare_models(sort= 'RMSE')
```

```
gbr_regr = create_model('gbr', fold=5)
```

```
tuned_regr = tune_model(gbr_regr,fold=5)
```

```
tuned_regr
```

```
rf_regr = create_model('rf', fold=5)
```

```
tuned_rf_regr = tune_model(rf_regr,fold=5)
```
